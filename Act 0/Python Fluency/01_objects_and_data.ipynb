### 1.1 — Identity vs equality

`==` compares **value**; `is` compares **identity** (same object in memory). Return a tuple 
`(equal, same_object)` for two inputs.

```
check([1,2], [1,2]) → (True, False)    # equal contents, different objects
```

````
def check(a, b):
    return a == b, a is b

# Tests
x = [1, 2]
assert check([1, 2], [1, 2]) == (True, False)
assert check(x, x) == (True, True)
assert check(1, 1) == (True, True)
print("1.1 passed")
````

### 1.2 — Aliasing in action

Append `value` to `lst`, then return the list. Demonstrate you understand that mutating a list 
through one name affects all names pointing to it: the test checks the *original* changed too.

````
def append_in_place(lst, value):
    lst.append(value)
    return lst

# Tests
original = [1, 2]
result = append_in_place(original, 3)
assert result == [1, 2, 3]
assert original == [1, 2, 3]   # same object — original changed
assert result is original
print("1.2 passed")
````

### 1.3 — Shallow vs deep copy

Given a list of lists, return a **deep** copy — modifying an inner list of the copy must NOT affect 
the original. Do it without `import copy`: build new inner lists yourself.

```
deep_copy([[1,2],[3]]) → [[1,2],[3]]  (fully independent)
```

````
def deep_copy(nested):
    deep = []
    for sublist in nested:
        subdeep = []
        for item in sublist:
            subdeep.append(item)
        deep.append(subdeep)

    return deep

# Tests
original = [[1, 2], [3, 4]]
copied = deep_copy(original)
copied[0].append(99)
assert original == [[1, 2], [3, 4]]   # untouched
assert copied == [[1, 2, 99], [3, 4]]
print("1.3 passed")
````

### 1.4 — Starred unpacking

Given a list with at least one element, return `(head, tail)` where `head` is the first element 
and `tail` is a list of the rest.

```
split_head([1,2,3,4]) → (1, [2,3,4])
split_head([5]) → (5, [])
```

````
def split_head(lst):
    head, *tail = lst
    return head, tail

# Tests
assert split_head([1, 2, 3, 4]) == (1, [2, 3, 4])
assert split_head([5]) == (5, [])
print("1.4 passed")
````

### 1.5 — Swap without a temp

Given a dict and two keys, swap their values in place and return the dict. Use tuple unpacking, 
no temporary variable.

```
swap({"a":1,"b":2}, "a", "b") → {"a":2,"b":1}
```

````
def swap(d, k1, k2):
    d[k1], d[k2] = d[k2], d[k1]
    return d

# Tests
assert swap({"a": 1, "b": 2}, "a", "b") == {"a": 2, "b": 1}
assert swap({"x": 10, "y": 20, "z": 30}, "x", "z") == {"x": 30, "y": 20, "z": 10}
print("1.5 passed")
````

### 1.6 — Why mutable things can't be keys

Return a set of the *hashable* items from a mixed list — skip anything that can't go in a set 
(lists, dicts). Use try/except around adding each item.

```
hashables([1, "a", [2,3], (4,5), {1:2}]) → {1, "a", (4,5)}
```

````
def hashables(items):
    returned_set = set()

    for item in items:
        try:
            returned_set.add(item)
        except TypeError:
            continue

    return returned_set


# Tests
assert hashables([1, "a", [2, 3], (4, 5), {1: 2}]) == {1, "a", (4, 5)}
assert hashables([[1], [2]]) == set()
print("1.6 passed")
````
