## 0.13 — Numbers and operators

The arithmetic operators, including the two that surprise people:

```python
7 + 3     # 10
7 - 3     # 4
7 * 3     # 21
7 / 3     # 2.333...  → TRUE division, always returns a float
7 // 3    # 2         → FLOOR division, rounds down to an int
7 % 3     # 1         → MODULO, the remainder
2 ** 10   # 1024      → exponentiation
```

The `/` vs `//` distinction matters enormously in this project. Array indices must be integers, 
so `len(lst) / 2` (a float) will break indexing — you want `len(lst) // 2`.

`%` (modulo) is how you do "every Nth" logic and wrap-around: `i % 2 == 0` tests for even, 
`i % n` cycles through `0, 1, ..., n-1`.

Negative floor division rounds **toward negative infinity**, not toward zero: `-7 // 2 == -4`, not `-3`.

## 0.14 — Boolean logic and chained comparisons

`and`, `or`, `not` — and they **short-circuit**:

```python
True and False    # False
True or False     # True
not True          # False
```

Short-circuit means evaluation stops as soon as the answer is known. `a and b` doesn't evaluate `b` 
if `a` is falsy. `a or b` doesn't evaluate `b` if `a` is truthy. This lets you write safe guards:

```python
if lst and lst[0] == 5:    # lst[0] only checked if lst is non-empty — no IndexError
```

Python also lets you **chain comparisons** the way math does:

```python
0 <= i < len(lst)     # reads exactly like the math: i is in [0, len)
1 < x < 100
```

This is cleaner than `i >= 0 and i < len(lst)` and means the same thing.

## 0.15 — f-strings

The modern way to build strings with values in them. Prefix the string with `f` and put 
expressions in `{}`:

```python
name = "Mara"
turns = 12
f"{name} held for {turns} turns"     # "Mara held for 12 turns"
f"{turns * 2} total"                 # "24 total" — any expression works
```

Number formatting goes after a colon inside the braces:

```python
loss = 3.14159265
f"{loss:.2f}"      # "3.14"   — 2 decimal places
f"{loss:.4f}"      # "3.1416"
f"{1000000:,}"     # "1,000,000"  — thousands separator
f"{0.875:.1%}"     # "87.5%"  — as a percentage
```

You'll use `f"{loss:.4f}"` constantly when printing training progress.

## 0.16 — List methods, and in-place vs returning

The methods, and a crucial distinction:

```python
lst.append(x)      # add ONE element to the end
lst.extend([a, b]) # add EACH element of an iterable to the end
lst.insert(i, x)   # insert x at position i
lst.pop()          # remove & return last element
lst.pop(i)         # remove & return element at i
lst.remove(x)      # remove first occurrence of value x
lst.index(x)       # position of first occurrence of x
lst.count(x)       # how many times x appears
```

`append` vs `extend` is the classic mix-up:

```python
a = [1, 2]
a.append([3, 4])   # [1, 2, [3, 4]]  — one element, which is a list
b = [1, 2]
b.extend([3, 4])   # [1, 2, 3, 4]    — two elements added
```

The other key distinction — **methods that change in place return `None`**:

```python
result = [3, 1, 2].sort()    # result is None! sort() mutates and returns nothing
```

`.sort()` reorders the list in place and returns `None`. If you want a **new** sorted list and 
leave the original alone, use the built-in `sorted()` instead:

```python
sorted([3, 1, 2])    # returns [1, 2, 3], original untouched
```

Same pattern: `.reverse()` (in place, returns None) vs `reversed()` / `[::-1]` (new). This is the 
same trap as your missing `return` — an in-place method gives you `None` if you try to capture it.

## 0.17 — `in`, and why your data structure choice matters

The `in` operator tests membership across all the container types:

```python
3 in [1, 2, 3]          # True
"a" in {"a", "b"}       # True
"key" in {"key": 1}     # True  — checks KEYS for dicts
"ell" in "hello"        # True  — substring check for strings
```

But here's the part that matters for performance: **checking `in` on a list is slow** — Python 
scans every element, so it's O(n). Checking `in` on a **set** or **dict** is O(1) — effectively instant, 
because they're hash-based.

This is exactly the inefficiency from your vocab-building code earlier:

```python
if character not in vocab:   # if vocab is a LIST, this scans the whole list every time
    vocab.append(character)
```

If `vocab` is a **set**, that membership check is instant — which is one reason a set was the right 
structure for collecting unique characters.

Rule of thumb: if you're repeatedly asking "is this thing in here?", use a set or dict, not a list.

## 0.18 — `sorted`, `min`, `max` with a `key`

`sorted`, `min`, and `max` all accept a `key` argument — a function that says *what to compare by*. 
This is one of the most useful patterns in Python.

```python
words = ["bb", "a", "cccc", "dd"]
sorted(words, key=len)              # ["a", "bb", "dd", "cccc"]  — by length
max(words, key=len)                 # "cccc"  — longest
min(words, key=len)                 # "a"     — shortest
```

The `key` function is applied to each element to get the value to sort/compare by. With `reverse=True` 
you flip the order:

```python
sorted([3, 1, 2], reverse=True)     # [3, 2, 1]
```

This is how you find the most frequent pair in BPE: `max(counts, key=counts.get)` returns the 
**key** whose **value** is largest. And it's how top-k sampling ranks tokens by probability.

## 0.20 — Tuples and immutability

A tuple is like a list but **immutable** — once created, you can't change its contents:

```python
t = (1, 2, 3)
t[0]            # 1   — indexing works
t[0] = 99       # TypeError! — can't assign
```

You create them with commas (parentheses are usually optional): `t = 1, 2` makes `(1, 2)`.

Why they matter: **immutable things can be dictionary keys and set elements; mutable things cannot.**

```python
d = {(1, 2): "a"}       # fine — tuple key
d = {[1, 2]: "a"}       # TypeError! — list can't be a key
```

This is exactly why pairs in BPE are tuples, not lists — you use them as dict keys in your 
pair-count map: `counts[(id_a, id_b)] += 1`. A list there would crash.

Tuples are also how functions return "multiple values" — `return a, b` packs them into a tuple 
the caller unpacks.

## 0.21 — Set operations


## 0.22 — Functions: defaults, multiple returns, the missing-`return` trap

**Default arguments** let a parameter be optional:

```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}"

greet("Mara")               # "Hello, Mara"
greet("Mara", "Evening")    # "Evening, Mara"
```

**Returning multiple values** — just separate with commas; the caller unpacks:

```python
def min_max(nums):
    return min(nums), max(nums)

lo, hi = min_max([3, 1, 4])     # lo=1, hi=4
```

**The missing-`return` trap** — the bug you hit several times today. A function with no `return` 
(or a `return` you forgot to reach) gives back `None`:

```python
def add(a, b):
    result = a + b      # computed...
                        # ...but never returned!

x = add(2, 3)           # x is None, not 5
```

If a function is supposed to produce a value, **it must `return` that value.** Computing it and 
leaving it in a local variable does nothing for the caller — that variable vanishes when the 
function ends. This is the single most common bug you've made; burn it in.

One trap to know about but avoid: **never use a mutable default** like `def f(x, items=[])`. The 
list is created once and shared across all calls. Use `def f(x, items=None)` and create it inside.

## 0.23 — `try` / `except` basics

When an operation might fail, you can catch the error instead of crashing:

```python
try:
    value = d[key]
except KeyError:
    value = "default"
```

The `try` block runs; if the named exception occurs, the `except` block runs instead of the program 
crashing. Common exceptions you'll meet:

- `KeyError` — dict key doesn't exist (you hit this earlier)
- `IndexError` — list index out of range
- `ValueError` — right type, wrong value (e.g. `int("abc")`)
- `TypeError` — wrong type (e.g. `set(5)`)
- `ZeroDivisionError` — divide by zero

For the specific "dict might not have this key" case, `.get(key, default)` is cleaner than 
try/except. Reach for try/except when there's no built-in guard, or when the failure is genuinely 
exceptional rather than expected. Don't wrap everything in try/except to hide bugs — let errors you 
didn't expect surface so you can fix them.

## 0.24 — Nested data structures

Real data is nested — lists of dicts, dicts of lists, dicts of dicts. You access deep into them by 
chaining indexing/keys:

```python
data = {"users": [{"name": "ann", "score": 90}, {"name": "bob", "score": 75}]}

data["users"]              # the list of dicts
data["users"][0]           # {"name": "ann", "score": 90}
data["users"][0]["name"]   # "ann"
```

Each bracket peels one layer. Read left to right: "in `data`, the `users` key, the 0th element, 
its `name` key."

You'll structure Mara's memory and config exactly like this — a config dict where 
`config["attention"]["type"]` is `"sliding_window"`, a memory record that's a dict of lists. 
Building and reading nested structures fluently is non-negotiable for the back half of the project.
