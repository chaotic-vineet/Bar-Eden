### 1.3 — Sliding windows

Given a list and a window size `k`, return all contiguous sublists of length `k`.

```
[1, 2, 3, 4, 5], k=3 → [[1, 2, 3], [2, 3, 4], [3, 4, 5]]
```

This pattern shows up in n-gram models and sliding window attention.

````python
def windows(lst, k):
    windowed = []

    for i in range(len(lst)-k+1):
        windowed.append(lst[i:i+k])

    return windowed

# Tests
assert windows([1, 2, 3, 4, 5], 3) == [[1, 2, 3], [2, 3, 4], [3, 4, 5]]
assert windows([1, 2, 3], 1) == [[1], [2], [3]]
assert windows([1, 2], 3) == []
assert windows([1, 2, 3, 4], 2) == [[1, 2], [2, 3], [3, 4]]
print("1.3 passed")
````

### 1.4 — Replace all occurrences of a pair

Given a list of integers, a pair `(a, b)`, and a replacement value `c`, 
replace every adjacent occurrence of `a, b` with `c`. Scan left to right —
after a replacement, skip past it.

```
[1, 2, 1, 2, 3], pair=(1,2), replacement=99 → [99, 99, 3]
[1, 1, 1], pair=(1,1), replacement=2 → [2, 1]
```

This is the core merge operation in BPE.

````python
def replace_pair(ids, pair, replacement):
    i=0
    while i < len(ids)-1:
        if ids[i] == pair[0] and ids[i+1] == pair[1]:
            ids[i] = replacement
            ids.pop(i+1)
        i += 1

# Tests
assert replace_pair([1, 2, 1, 2, 3], (1, 2), 99) == [99, 99, 3]
assert replace_pair([1, 1, 1], (1, 1), 2) == [2, 1]
assert replace_pair([1, 1, 1, 1], (1, 1), 2) == [2, 2]
assert replace_pair([3, 4, 5], (1, 2), 99) == [3, 4, 5]
assert replace_pair([], (1, 2), 99) == []
print("1.4 passed")
````
