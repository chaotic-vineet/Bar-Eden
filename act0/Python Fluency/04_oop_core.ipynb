### 4.1 — Instance attributes and methods

Build a `BankAccount` with a starting `balance`, a `deposit(amount)` method, and a 
`withdraw(amount)` method that refuses to overdraw (returns `False` and leaves balance unchanged 
if insufficient; returns `True` on success).

```
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            return True
        return False

# Tests
a = BankAccount(100)
a.deposit(50)
assert a.balance == 150
assert a.withdraw(30) == True
assert a.balance == 120
assert a.withdraw(1000) == False
assert a.balance == 120
print("4.1 passed")
```

### 4.2 — Class attributes vs instance attributes

A **class attribute** is shared by all instances; an **instance attribute** belongs to one object. 
A class attribute is a clean way to count how many instances exist.

Build a `Widget` class where every instance increments a shared `count`, accessible as `Widget.count`.

```
class Widget:
    count = 0   # class attribute — shared

    def __init__(self):
        Widget.count += 1

# Tests
Widget.count = 0
a = Widget()
b = Widget()
c = Widget()
assert Widget.count == 3
print("4.2 passed")
```

### 4.3 — `@classmethod` and `@staticmethod`

- An **instance method** takes `self`.
- A **`@classmethod`** takes `cls` (the class) — useful for alternative constructors.
- A **`@staticmethod`** takes neither — it's a plain function namespaced under the class.

Build `Temperature` storing celsius, with a classmethod `from_fahrenheit(f)` that constructs an 
instance, and a staticmethod `is_freezing(c)` returning whether a celsius value is at or below 0.

```
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    @classmethod
    def from_fahrenheit(cls, f):
        c = (f-32)*5/9
        return cls(c)

    @staticmethod
    def is_freezing(c):
        if c <= 0:
            return True
        return False

# Tests
t = Temperature.from_fahrenheit(32)
assert abs(t.celsius - 0) < 1e-9
assert isinstance(t, Temperature)
assert Temperature.is_freezing(-5) == True
assert Temperature.is_freezing(10) == False
print("4.3 passed")
```

### 4.4 — `@property`

A `@property` lets a method be accessed like an attribute — the Pythonic way to expose a computed 
value or guard access, without callers writing `()`.

Build `Circle` with a `radius`, and a `area` property that returns the area computed on access.

```
class Circle:
    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):
        area = 3.14159*(self.radius**2)
        return area

# Tests
c = Circle(2)
assert abs(c.area - 12.56636) < 1e-5    # accessed as attribute, no ()
c.radius = 3
assert abs(c.area - 28.27431) < 1e-5    # recomputed
print("4.4 passed")
```
